Запуск моделей

```python
from sklearn.linear_model import LassoCV
lasso_cv = Lasso(max_iter=50000, alpha=0.015)
# lasso_cv = LassoCV(cv=5)
lasso_cv.fit(X_train, y_train)
# lasso_cv.coef_
```

Трансформер

```python
# Analyze selected features and their importance
sfm = SelectFromModel(estimator=lasso_cv, prefit=True)
selected_feature_indices = np.where(sfm.get_support())[0]
selected_features = features.columns[selected_feature_indices]
coefficients = lasso_cv.coef_
print("Selected Features:", selected_features)
# print("Feature Coefficients:", coefficients)
print(len(selected_features))
```

```python

ratio_transformer = Pipeline(
    steps=[
          ('imputer', KNNImputer(n_neighbors=3)),
        ('scaler', RobustScaler())
    ]
)


column_transformer = ColumnTransformer([
    ('r',ratio_transformer, selected_features )
]
)
column_transformer.set_output(transform='pandas') # вот такой трансформер получился
column_transformer
```

```python
from sklearn.ensemble import GradientBoostingRegressor
clf = RandomForestRegressor(n_estimators = 500)
# clf = GradientBoostingRegressor(random_state=0, n_estimators=2000)
pipe = Pipeline(steps=[
    # ('transform', column_transformer),
    ('selector', sfm),
    ('clf', clf)
])
```

```python
%time

pipe.fit(X_train, y_train)

predicted = pipe.predict(X_test)
```

```python
rmse = mean_squared_error(y_test, predicted, squared=False)
print('rmse =  ', rmse) # бейзлайн построен и он имеет вот такую ошибку.
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predicted)
print('r2_score =  ', r2) # бейзлайн построен и он имеет вот такую ошибку.
# дальше будем проверять работу других моделей и думать, как и что улучшить.
```

```python
import numpy as np
from catboost import Pool, CatBoostRegressor

# initialize Pool



train_pool = Pool(column_transformer.fit_transform(X_train[selected_features]),
                  y_train,
                  )
test_pool = Pool(column_transformer.transform(X_test[selected_features]),
                 )

# specify the training parameters

model = CatBoostRegressor(iterations=4000,
                          depth=4,
                          learning_rate=0.015,
                          loss_function='RMSE',
                          early_stopping_rounds=50
                          , verbose=1000)
#train the model
model.fit(train_pool)
# make the prediction using the resulting model
predicted = model.predict(test_pool)
rmse = mean_squared_error(y_test, predicted, squared=False)
print('rmse =  ', rmse) # бейзлайн построен и он имеет вот такую ошибку.
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predicted)
print('r2_score =  ', r2) # бейзлайн построен и он имеет вот такую ошибку.
# дальше будем проверять работу других моделей и думать, как и что улучшить.
```
